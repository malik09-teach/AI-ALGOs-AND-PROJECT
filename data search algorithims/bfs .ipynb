```python
import networkx as nx

G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F')])

start_node = 'A'

bfs_edges = list(nx.bfs_edges(G, source=start_node))

bfs_nodes = [start_node] + [v for u, v in bfs_edges]

print(f"BFS Node Order: {bfs_nodes}")
```

```python
import matplotlib.pyplot as plt

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=800, font_weight='bold')
plt.show()
```
